```python
tabla_p = pd.crosstab(index=data['Estrato'],
            columns=data['Opinión'], margins=True)
tabla_p
```

```python
# CANTIDAD DE INDEX Y COLUMNS EN UNA TABLA
# ESTRATO             OPINIÓN
f = list(tabla_p.columns)
g = tabla_p.index

print(g)
print(len(g)-1)

print(f)
print(len(f)-1)
list(range(4))
```

```python
class EstaGamma:
    
    def __init__(self):
        pass
    
    def __str__(self):
        return("Ingresar en el metodo concordantes y discordantes (g,f) las filas en g y las columnas en f ,Ingresar en el metodo gamma() sin ningun atributo") 
    
    def concordantes(self,g,f):
        global c
        c=0
        k = list(range(len(g)-1)) # SI SE GENERALIZA DE ALGUNA FORMA MEJOR
        l = list(range(len(f)-1))  
        for i in range(len(g)-1): # filas --> 4
            for j in range(len(f)-1): # columnas --> 3
                s = 0
                print("------------------------------------")
                print("Estamos en el par (",i,",",j ,")y el valor de este es :",tabla_p.iloc[i,j])
                elemento = tabla_p.iloc[i,j]
                for x in k :
                    for y in l :
                        if  x > i:
                            if y > j :
                                print("El par ",(x,y)," es :",tabla_p.iloc[x,y])
                                s = s + tabla_p.iloc[x,y]
                print("La Suma es de todos los pares hallados es ",s)
                print("El Producto seria ",s,"con ",elemento,"que resulta ->",s*elemento)
                c = c + elemento*s
         
        
        return c
        
    def discordantes(self,g,f):
        global d
        d = 0
        k = list(range(len(g)-1)) ##SI SE GENERALIZA DE ALGUNA FORMA MEJOR
        l = list(range(len(f)-1))
        for i in range(len(g)-1): # filas --> 4
            for j in range(len(f)-1): # columnas --> 3
                s = 0
                print("--------------------------------------")
                print("Estamos en el par (",i,",",j,")y el valor de este es :",tabla_p.iloc[i,j])
                elemento = tabla_p.iloc[i,j]
                for x in k :
                    for y in l :
                        if  x > i:
                            if y < j :
                                print("El par ",(x,y)," es :",tabla_p.iloc[x,y])
                                s = s + tabla_p.iloc[x,y]
                print("La Suma es ",s)
                print("El Producto seria ",s,"con ",elemento,"que resulta ->",s*elemento)
                d = d + elemento*s
                
       
        return d
    
    def gamma(self):
        eg = (c-d)/(c+d)
        print("EL ESTADISTICO GAMMA PARA LA TABLA PROPUESTA ES",eg)
        

    
    
```

```python
Ola= EstaGamma()
print(Ola)
Ola.concordantes(g,f)
Ola.discordantes(g,f)
Ola.gamma()
```
